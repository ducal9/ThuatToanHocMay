```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import mglearn

# generate some random cluster data
X, y = make_blobs(random_state=170, n_samples=600)
rng = np.random.RandomState(74)

# transform the data to be stretched
transformation = rng.normal(size=(2, 2))
X = np.dot(X, transformation)

# cluster the data into three clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=mglearn.cm3)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='^', c=[0, 1, 2], s=100, linewidth=2, cmap=mglearn.cm3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons
import mglearn

# generate synthetic two_moons data (with less noise this time)
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# cluster the data into two clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=mglearn.cm2, s=60)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='^', c=[mglearn.cm2(0), mglearn.cm2(1)], s=100, linewidth=2)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```


import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF, PCA
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people

# 1) Tải dữ liệu khuôn mặt (LFW) và chuẩn bị X_people, y_people
people = fetch_lfw_people(min_faces_per_person=20, resize=0.4, color=False)
X_people = people.data.astype(np.float32)   # shape (n_samples, n_features)
y_people = people.target
image_shape = people.images[0].shape        # để vẽ lại ảnh

# 2) Đưa dữ liệu về [0,1] để phù hợp NMF (không âm)
X_people /= 255.0

# 3) Chia train/test có stratify
X_train, X_test, y_train, y_test = train_test_split(
    X_people, y_people, stratify=y_people, random_state=0
)

# 4) Khởi tạo & huấn luyện các mô hình nén/biểu diễn
nmf = NMF(n_components=100, random_state=0, init="nndsvda", max_iter=400)
nmf.fit(X_train)

pca = PCA(n_components=100, random_state=0, whiten=False)
pca.fit(X_train)

kmeans = KMeans(n_clusters=100, random_state=0, n_init="auto")
kmeans.fit(X_train)

# 5) Tái tạo ảnh từ các mô hình
X_reconstructed_pca = pca.inverse_transform(pca.transform(X_test))
X_reconstructed_kmeans = kmeans.cluster_centers_[kmeans.predict(X_test)]
# Có thể dùng nmf.inverse_transform, hoặc giữ phép nhân như bạn đang dùng:
X_reconstructed_nmf = nmf.inverse_transform(nmf.transform(X_test))
# (tương đương với: np.dot(nmf.transform(X_test), nmf.components_))

# 6) Hiển thị so sánh 6 ảnh: gốc | PCA | KMeans | NMF
n_show = 6
plt.figure(figsize=(10, 7))
for i in range(n_show):
    # Original
    ax = plt.subplot(4, n_show, i + 1)
    plt.imshow(X_test[i].reshape(image_shape), cmap="gray")
    ax.set_title("Original")
    plt.axis("off")

    # PCA
    ax = plt.subplot(4, n_show, i + 1 + n_show)
    plt.imshow(np.clip(X_reconstructed_pca[i], 0, 1).reshape(image_shape), cmap="gray")
    ax.set_title("PCA")
    plt.axis("off")

    # KMeans
    ax = plt.subplot(4, n_show, i + 1 + 2*n_show)
    plt.imshow(np.clip(X_reconstructed_kmeans[i], 0, 1).reshape(image_shape), cmap="gray")
    ax.set_title("KMeans")
    plt.axis("off")

    # NMF
    ax = plt.subplot(4, n_show, i + 1 + 3*n_show)
    plt.imshow(np.clip(X_reconstructed_nmf[i], 0, 1).reshape(image_shape), cmap="gray")
    ax.set_title("NMF")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```
fig, axes = plt.subplots(3, 5, figsize=(8, 8),
 subplot_kw={'xticks': (), 'yticks': ()})
fig.suptitle("Extracted Components")
for ax, comp_kmeans, comp_pca, comp_nmf in zip(
 axes.T, kmeans.cluster_centers_, pca.components_, nmf.components_):
 ax[0].imshow(comp_kmeans.reshape(image_shape))
 ax[1].imshow(comp_pca.reshape(image_shape), cmap='viridis')
 ax[2].imshow(comp_nmf.reshape(image_shape))
axes[0, 0].set_ylabel("kmeans")
axes[1, 0].set_ylabel("pca")
axes[2, 0].set_ylabel("nmf")
fig, axes = plt.subplots(4, 5, subplot_kw={'xticks': (), 'yticks': ()},
 figsize=(8, 8))
fig.suptitle("Reconstructions")
for ax, orig, rec_kmeans, rec_pca, rec_nmf in zip(
 axes.T, X_test, X_reconstructed_kmeans, X_reconstructed_pca,
 X_reconstructed_nmf):
 ax[0].imshow(orig.reshape(image_shape))
 ax[1].imshow(rec_kmeans.reshape(image_shape))
 ax[2].imshow(rec_pca.reshape(image_shape))
 ax[3].imshow(rec_nmf.reshape(image_shape))
axes[0, 0].set_ylabel("original")
axes[1, 0].set_ylabel("kmeans")
axes[2, 0].set_ylabel("pca")
axes[3, 0].set_ylabel("nmf")
```

```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, ward

# Tạo dữ liệu mẫu
X, y = make_blobs(random_state=0, n_samples=12)

# Thực hiện phân cụm phân cấp (Ward linkage)
linkage_array = ward(X)

# Vẽ sơ đồ dendrogram
plt.figure(figsize=(8, 4))
dendrogram(linkage_array)

# Đánh dấu các mức cắt (2 cụm và 3 cụm)
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [7.25, 7.25], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')
ax.text(bounds[1], 7.25, ' two clusters', va='center', fontdict={'size': 12})
ax.text(bounds[1], 4, ' three clusters', va='center', fontdict={'size': 12})

plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.title("Dendrogram - Ward Hierarchical Clustering")
plt.show()
```

```
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import mglearn

# Tạo dữ liệu hình "hai vầng trăng"
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Chuẩn hóa dữ liệu về trung bình 0 và phương sai 1
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

# Khởi tạo và huấn luyện mô hình DBSCAN
# dbscan = DBSCAN()
dbscan = DBSCAN(eps=0.5, min_samples=5)

clusters = dbscan.fit_predict(X_scaled)

# Vẽ biểu đồ phân cụm
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm2, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.title("DBSCAN Clustering on Two Moons")
plt.show()
```
