```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = df.copy()  # To bn sao ca df  gi nguyn d liu gc
df_scaled[['GK', 'Buoi_hoc', 'Diem_BT']] = scaler.fit_transform(df[['GK', 'Buoi_hoc', 'Diem_BT']])
```

```python
scaler_X = scaler.fit_transform(df)
```

```python
scaler_X
```

```python
df_scaled[['GK', 'Buoi_hoc', 'Diem_BT']]
```

```python
from sklearn.cluster import DBSCAN
```

```python
number_of_outliers = []
percentage_of_outliers = []

# min_samples starting point: 2 times the number of dimensions
min_samples = 2 * df_scaled[['GK', 'Buoi_hoc', 'Diem_BT']].shape[1]

# Iterate over a range of epsilon values
for eps in np.linspace(0.001, 3, 50):
    # Create and fit the DBSCAN model
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan.fit(df_scaled[['GK', 'Buoi_hoc', 'Diem_BT']])
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    # Percentage of points classified as outliers
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    percentage_of_outliers.append(perc_outliers)
```

```python
number_of_outliers = []
percentage_of_outliers = []

# min_samples starting point: 2 times the number of dimensions
min_samples = 2 * scaler_X.shape[1]

# Iterate over a range of epsilon values
for eps in np.linspace(0.001, 3, 50):
    # Create and fit the DBSCAN model
    dbscan = DBSCAN(eps=eps, min_samples=1)
    dbscan.fit(scaler_X)
    
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    # Percentage of points classified as outliers
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    percentage_of_outliers.append(perc_outliers)
```

```python
sns.lineplot(x = np.linspace(0.01, 3, 50)  ,y= percentage_of_outliers);
plt.xlabel('Epsilon Value')
plt.ylabel('percentage_of_outliers');
```
